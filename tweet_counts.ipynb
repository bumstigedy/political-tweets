```python
import psycopg2
import pandas as pd
```

```python
conn = psycopg2.connect(
    host="localhost",
    database="capstone",
    user="postgres",
    password="postgres")
```

```python
sql="""SELECT text from tweets where text like '@%'"""
```

```python
df=pd.read_sql(sql,conn)
```

```python
df.head()
```

```python
import numpy as np
```

```python
who_replies_to=[]
for tweet in df.text:
    parts = tweet.split()
    for i in range(0,len(parts)):
            if parts[i][0]=='@':
                who_replies_to.append(parts[i])
                
```

```python
def tweet_counts(tweet_array,threshold=10):
    unique,counts = np.unique(np.array(tweet_array),return_counts=True)
    count_dict=dict(zip(unique,counts))
    return [(k,v) for k,v in count_dict.items() if v > threshold ]
```

```python
tweet_counts(who_replies_to,5)
```

```python
replies_df=pd.DataFrame(tweet_counts(who_replies_to,5),columns=['who','count']).sort_values('count',ascending=False)
```
